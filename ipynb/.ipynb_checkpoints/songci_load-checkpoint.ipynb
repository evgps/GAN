```python
def readsongci(filename):
    db = sql.connect(filename)
    data = db.cursor()
    ci = data.execute("select * from " + re.split(r'[\/\.]',filename)[-2])
    songci = []
    for seq in ci:
        songci.append(seq[3].strip().split('\n'))
    return songci
```

```python
def ciandpoet2seq(poets):
    data = []
    for p in poets:
        for seq in p:
            data.append([c for c in jieba.cut(seq)])
    return data    
```
